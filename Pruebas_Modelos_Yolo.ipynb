```python
# ============================================================
#   IMPORTACIÓN DE LIBRERÍAS
# ============================================================
from ultralytics import YOLO   # Permite cargar y ejecutar modelos YOLO
import torch                   # Se usa para verificar entorno PyTorch

# Se imprime la versión de PyTorch y si hay GPU disponible
print("PyTorch version:", torch.__version__)
print("GPU disponible:", torch.cuda.is_available())
```

```python
# ============================================================
#   CARGA DEL MODELO DE DETECCIÓN DE BASURA
# ============================================================
# El usuario sube el archivo .pt con el modelo entrenado (best_basura.pt)

from google.colab import files

uploaded = files.upload()  # Subir best.pt
```

```python
model = YOLO("best_basura.pt")  ## Se carga el modelo YOLO utilizando el nombre del archivo subido
model
```

```python
# ============================================================
#   SUBIR UNA IMAGEN Y REALIZAR LA PREDICCIÓN (BASURA)
# ============================================================
# Se sube la imagen que se desea analizar
from google.colab import files
uploaded_image = files.upload()

# Se obtiene la ruta del archivo subido (el primer archivo seleccionado)
image_path = list(uploaded_image.keys())[0]
image_path # Muestra la ruta interna de Colab
```

```python
model.names
```

```python
# Se realiza la predicción usando el modelo de basura
results = model.predict(source=image_path)
# Se muestra la imagen con las detecciones (cajas, etiquetas, etc.)
results[0].show()
```
