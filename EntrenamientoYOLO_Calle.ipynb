```python

# ============================================================
#   DESCARGA DEL DATASET DESDE KAGGLE
# ============================================================
import kagglehub


# Descarga automÃ¡tica del dataset pÃºblico de detecciÃ³n de huecos
# Dataset: rajdalsaniya/pothole-detection-dataset
# kagglehub descarga el dataset directamente a la carpeta de cachÃ© del sistema.
path = kagglehub.dataset_download("rajdalsaniya/pothole-detection-dataset")
print("Dataset descargado en:", path)
```

```python
# ============================================================
#   EXPLORACIÃ“N DEL DATASET DESCARGADO
# ============================================================
# Se recorre la estructura de carpetas del dataset para visualizar su contenido.
import os

for root, dirs, files in os.walk(path):
    print(root)
```

```python
# ============================================================
#   CREACIÃ“N DEL ARCHIVO data.yaml PARA YOLOv8
# ============================================================
# YOLO requiere un archivo YAML donde se especifica:
# - Ruta del dataset
# - Carpeta de entrenamiento
# - Carpeta de validaciÃ³n
# - NÃºmero de clases
# - Nombre de cada clase

%%writefile /content/data.yaml
path: "/root/.cache/kagglehub/datasets/rajdalsaniya/pothole-detection-dataset/versions/1"
train: "train/images"
val: "valid/images"

nc: 1
names: ["pothole"]
```

```python
# ============================================================
#   REVISAR QUE LAS RUTAS DEL DATASET SEAN CORRECTAS
# ============================================================
import os

base = "/root/.cache/kagglehub/datasets/rajdalsaniya/pothole-detection-dataset/versions/1"
# Se listan las carpetas principales del dataset
print(os.listdir(base))

# Se verifican las carpetas de imÃ¡genes y anotaciones de train/valid
print("Train:", os.listdir(base + "/train"))
print("Valid:", os.listdir(base + "/valid"))
```

```python
# ============================================================
#   CARGA DEL MODELO BASE YOLOv8 PARA ENTRENAMIENTO
# ============================================================
from ultralytics import YOLO

# Se carga el modelo base YOLOv8 Medium (yolov8m.pt)
# Este modelo serÃ¡ reentrenado (fine-tuning) con nuestro dataset.
model = YOLO("yolov8m.pt")

# ============================================================
#   ENTRENAMIENTO DEL MODELO
# ============================================================
# model.train() entrena el modelo con los parÃ¡metros especificados:
# - data="data.yaml": archivo que describe el dataset
# - epochs=50: nÃºmero de Ã©pocas
# - imgsz=640: tamaÃ±o de las imÃ¡genes de entrada
# - batch=16: tamaÃ±o del batch en cada iteraciÃ³n
# - patience=20: detiene el entrenamiento si no mejora
# - device=0: usa GPU (si estÃ¡ disponible)
# - name="pothole_yolov8": carpeta de salida donde se guardarÃ¡n los resultados
model.train(
    data="data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=20,
    device=0,  # usa GPU
    name="pothole_yolov8"
)
```
